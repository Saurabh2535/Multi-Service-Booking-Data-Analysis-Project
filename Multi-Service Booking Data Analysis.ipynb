***Checking Missing Values***

```python
df.isna().sum()
```

We find missing values in Class Type, Instructor, Time Slot, Duration, Facility, Theme, Customer Email, Subscription Type and Customer Phone.
Subscription Type column is completely empty (might be irrelevant). 
    So we will drop this column.

```python
df.drop(columns='Subscription Type',inplace=True)
df
```

***Filling Missing Values:***

Instructor → Fill with "Unknown"

Time Slot, Duration (mins) → Fill "Not Provided" for missing values

Facility, Theme → Fill "Not Applicable" for missing values

Customer Email, Customer Phone → Keep blank (may help identify missing customer data)

```python
df['Instructor'] = df['Instructor'].fillna("Unknown")
df[['Time Slot', 'Duration (mins)']] = df[['Time Slot', 'Duration (mins)']].fillna("Not Provided")
df[['Facility', 'Theme']] = df[['Facility', 'Theme']].fillna("Not Applicable")
```

```python
df
```

***Creating new columns Month and Weekday from Booking Date***

```python
df['Month Name'] = df['Booking Date'].dt.month_name()
df['Weekday Name'] = df['Booking Date'].dt.day_name()
```

```python
df
```

***Booking Types***

```python
print(df['Booking Type'].unique())
```

```python
# Count of unique values in 'Booking Type' column
booking_type_counts = df['Booking Type'].value_counts()

# Pie chart
plt.figure(figsize=(10, 6))
booking_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('% of Unique Values in Booking Type Column')
plt.ylabel('')
plt.show()
```

***Booking Status vs Booking Type***

```python
# Chart to visualize the relationship between Booking Status and Booking Type
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Booking Type', hue='Status', data=df)
plt.xlabel("Booking Type")
plt.ylabel("Count")
plt.title("Booking Status vs Booking Type")
for i in ax.containers:
    ax.bar_label(i)
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by Service Type and sum the revenue (Price column)
service_revenue = df.groupby('Service Name')['Price'].sum()

# Pie Chart
plt.figure(figsize=(8, 8))
colors = sns.color_palette("pastel")

plt.pie(service_revenue, labels=service_revenue.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Total Revenue by Service Type')

plt.show()
```

```python
# Chart to visualize booking trends on the basis of Month Name
plt.figure(figsize=(12, 6))
ax = df['Weekday Name'].value_counts().plot(kind='bar')
plt.xlabel("Day Name")
plt.ylabel("Number of Bookings")
plt.title("Booking Trends by Day Name")
for i in ax.containers:
    ax.bar_label(i)
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of bookings per Time Slot
time_slot_counts = df['Time Slot'].value_counts().reset_index()
time_slot_counts.columns = ['Time Slot', 'Booking Count']

#  Bar Chart: Most Booked Time Slots
plt.figure(figsize=(12, 6))
sns.barplot(y='Time Slot', x='Booking Count', data=time_slot_counts, palette='coolwarm')
plt.xlabel('Number of Bookings')
plt.ylabel('Time Slot')
plt.title('Most Booked Time Slots')
plt.xticks(rotation=45)
plt.show()

#  Heatmap: Booking Distribution by Time Slots
plt.figure(figsize=(10, 6))
time_slot_pivot = time_slot_counts.pivot_table(index='Time Slot', values='Booking Count', aggfunc='sum')
sns.heatmap(time_slot_pivot, annot=True, fmt=".0f", cmap="Blues", linewidths=0.5)
plt.title('Booking Distribution Heatmap')
plt.ylabel('Time Slot')
plt.show()
```
